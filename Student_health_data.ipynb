```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregando o dataset
# Objetivo: Extrair insights do dataset sobre a saúde dos estudantes
dataset = pd.read_csv('student_health_data.csv')
print(dataset.head())
# Relação estresse x atividade física
plt.figure(figsize=(10, 6))
sns.barplot(data = dataset, x='Stress_Level_Biosensor', y='Physical_Activity', palette = 'Reds_r', hue = 'Gender')
plt.title('Estresse x Atividade Fisica')
plt.xlabel('Nível de estresse biosensor')
plt.ylabel('Atividade Fisica')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(data = dataset, x='Stress_Level_Self_Report', y='Physical_Activity', palette = 'Reds_r', hue = 'Gender')
plt.title('Estresse x Atividade Fisica')
plt.xlabel('Nível de estresse percebido')
plt.ylabel('Atividade Fisica')
plt.show()

# Relação Estresse x Qualidade de sono
plt.figure(figsize=(10, 6))
sns.barplot(data = dataset, x='Stress_Level_Biosensor', y='Sleep_Quality', palette = 'pastel', hue = 'Gender')
plt.title('Estresse x Qualidade de Sono')
plt.xlabel('Nível de estresse biosensor')
plt.ylabel('Qualidade de Sono')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(data = dataset, x='Stress_Level_Self_Report', y='Sleep_Quality', palette = 'pastel', hue = 'Gender')
plt.title('Estresse x Qualidade de Sono')
plt.xlabel('Nível de estresse percebido')
plt.ylabel('Qualidade de Sono')
plt.show()

# Relação Estresse x Horas de estudo
bins = [0, 2, 4, 6, 8, 10]
labels = ['0-2h', '2-4h', '4-6h', '6-8h', '8-10h']
dataset['StudyGroup'] = pd.cut(dataset['Study_Hours'], bins = bins, labels = labels)

plt.figure(figsize=(10, 6))
sns.barplot(data = dataset, x='StudyGroup', y='Stress_Level_Biosensor', ci='sd', hue = 'Gender')
plt.title('Estresse x Horas de estudo')
plt.xlabel('Horas de Estudo')
plt.ylabel('Nível de estresse biosensor')
plt.legend(title = 'Genêro')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(data = dataset, x='StudyGroup', y='Stress_Level_Self_Report', ci='sd', hue = 'Gender')
plt.title('Estresse x Horas de estudo')
plt.xlabel('Horas de Estudo')
plt.ylabel('Nível de estresse percebido')
plt.legend(title = 'Genêro')
plt.show()
```
